```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("full_count.csv")
```

```python
df["average"].hist(bins=50)

# Add labels and title
plt.xlabel("Average")
plt.ylabel("Frequency")
plt.title("Histogram of 'AADT' column with 50 bins")

# Show plot
plt.show()
```

```python
print("Range: " , np.max(df["average"])-np.min(df["average"]) )
print("mean: " , np.mean(df["average"]) )
print("std: " , np.std(df["average"]) )
print("max: " , np.max(df["average"]) )
print("min: " , np.min(df["average"]) )
```
